```python
import matplotlib.pyplot as plt
```

#### Graph 1

Evolucion ICP en Espa√±a (https://www.msn.com/es-es/entretenimiento/tv/gr%C3%A1ficas-manipuladas-vs-la-realidad-indignaci%C3%B3n-por-otro-error-en-el-programa-de-ana-rosa-ahora-con-el-ipc/ar-AA13yXIW)

![alt text](ipc.jpg)

```python


# Datos
x = ["ENE","FEB","MAR","ABR","MAY","JUN","JUL","AGO","SEPT","OCT"]
y = [6.1,7.6,9.8,8.3,8.7,10.2,10.8,10.5,8.9,7.3]

# Crear la gr√°fica
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(x, y, color='#2D87BB', width=0.5)

# Agregar etiquetas de valor encima de las barras
for i, bar in enumerate(bars):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height+0.2, '%.1f' % height,
            ha='center', va='bottom', color='black', fontsize=12, fontweight='bold')

# Remover la escala en el eje y
ax.set_yticks([])

# Cambiar los colores de fondo y de los ejes
fig.set_facecolor('#E5E5E5')
ax.set_facecolor('#E5E5E5')
ax.spines['bottom'].set_color('gray')
ax.spines['top'].set_color('gray')
ax.spines['right'].set_color('gray')
ax.spines['left'].set_color('gray')

# A√±adir t√≠tulos y etiquetas de los ejes
plt.title("Evoluci√≥n del ICP", fontsize=20, fontweight='bold', color='#333333')
plt.xlabel("Meses", fontsize=14, fontweight='bold', color='#333333')

# Cambiar el tama√±o de la fuente de los ticks del eje x
plt.xticks(fontsize=12)
plt.ylim(0,12)
# Mostrar la gr√°fica
plt.show()
```

#### Graph 3
